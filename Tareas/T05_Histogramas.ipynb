## Histograms

```python
import pandas as pd

# Import the air quality dataset 
terminos = pd.read_csv("../data/Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(terminos.info())
```

Plot an histogram

```python
import matplotlib.pyplot as plt

# Select data
sal_data = terminos["sal_psu"]

plt.hist(sal_data, bins=10)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

plot histogram for specific variable with density

```python
# Plot the histogram with density (nomalized)
# Select data 
data = terminos["sal_psu"]


plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

Boxplots

```python
# simple boxplot

do_data = terminos["do_mg_l"]
plt.boxplot(do_data)

plt.xlabel("DO (mg/L)")
plt.ylabel("Concentration")
plt.title("Boxplot of DO Concentration")

# show plot
plt.show()
```
