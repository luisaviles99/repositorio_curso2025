```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.boxplot(x="habitat", y="individualCount", data= crust)
plt.title("crustacean counts by habitat")
#plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
sns.boxplot(x="locality", y="individualCount", data= crust)
plt.title("crustacean counts by locality")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show
```
